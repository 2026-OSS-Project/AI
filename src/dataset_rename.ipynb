```python
import os
import re

folder_path = "/content/drive/MyDrive/OSS/dataset/raw/vinyl"

files = os.listdir(folder_path)

# 기존 vinyl_번호 파일 찾기
vinyl_nums = []

for file in files:
    match = re.match(r"vinyl_(\d+)", file)
    if match:
        vinyl_nums.append(int(match.group(1)))

# 다음 번호 계산
start_num = max(vinyl_nums) + 1

# cardboard 파일만 추출
cardboard_files = sorted(
    [f for f in files if f.startswith("cardboard")]
)

# 이름 변경
for file in cardboard_files:

    ext = os.path.splitext(file)[1]

    new_name = f"vinyl_{start_num:03d}{ext}"

    old_path = os.path.join(folder_path, file)
    new_path = os.path.join(folder_path, new_name)

    os.rename(old_path, new_path)

    print(f"{file} -> {new_name}")

    start_num += 1

print("변경 완료")
```

```python
import os

folder_path = "/content/drive/MyDrive/OSS/dataset/raw/vinyl"

# jpg/jpeg만 추출
image_files = sorted([
    f for f in os.listdir(folder_path)
    if f.lower().endswith((".jpg", ".jpeg"))
])

print(f"대상 파일 수: {len(image_files)}")

# 1단계: 임시 이름으로 변경
for i, file in enumerate(image_files):

    old_path = os.path.join(folder_path, file)

    temp_name = f"temp_{i:05d}.jpg"

    temp_path = os.path.join(folder_path, temp_name)

    os.rename(old_path, temp_path)

# 2단계: vinyl_001.jpg 형식으로 변경
temp_files = sorted([
    f for f in os.listdir(folder_path)
    if f.startswith("temp_")
])

for i, file in enumerate(temp_files, start=1):

    old_path = os.path.join(folder_path, file)

    new_name = f"vinyl_{i:03d}.jpg"

    new_path = os.path.join(folder_path, new_name)

    os.rename(old_path, new_path)

print(f"{len(temp_files)}개 파일 이름 변경 완료")
```
